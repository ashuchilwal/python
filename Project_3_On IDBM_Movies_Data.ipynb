## Ans5  Find the movies directed by 'James Gunn', & 'Ridley Scott' with the rating greater than or equal to 8.

```python
mark1 = data['Director'] == 'James Gunn'
mark2 = data['Director'] == 'Ridley Scott'
mark3 = data['Rating'] >= 8

new_data = data.loc[((mark1 | mark2) & mark3), ['Title', 'Director', 'Rating']]
new_data
```

## Ans6 Display all the movies directed by 'Ridley Scott'.

## Ans7 Display all the director's who got the rating >= 8 for their movies
